#### 1. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.

#### 3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.

### Means 

```python
mean = dataset[excludedColumns].mean()
print(f'All Means \n{mean}')
```

### Standard Deviation

```python
standardDeviation = dataset[excludedColumns].std()
print(f'All Standard Deviations \n{standardDeviation}')
```

### Quartile Values

#### Quartile 1

```python
quartile1 = dataset[excludedColumns].quantile(0.25)
print(f'All quartile 1 \n{quartile1}')
```

#### Quartile 2

```python
quartile2 = dataset[excludedColumns].quantile(0.50)
print(f'All quartile 2 \n{quartile2}')
```

#### Quartile 3

```python
quartile3 = dataset[excludedColumns].quantile(0.75)
print(f'All quartile 3 \n{quartile3}')
```

#### All Quartile Values

```python
allQuartiles = dataset[excludedColumns].quantile([0.25,0.50,0.75])
print(allQuartiles)
```

```python
All = dataset[excludedColumns].describe()
print(All)
```

#### 4. Plot the histogram of each of these features in a 1x4 layout.

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
# fig -- the entire figure object (container for all subplots).
# axes -- an array of 4 subplot axes (1D array)

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 4))

for i in range(4):
    axes[i].hist(dataset[excludedColumns[i]], bins=30, color='Yellow', edgecolor='black') # bins=30 -- divides the data into 30 bins (bars).
    axes[i].set_title(f'Histogram {i+1}')
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel('Frequency')
```

```python
g = sn.PairGrid(dataset[excludedColumns])
g.map_upper(sn.histplot) #bivariate histogram
g.map_diag(sn.histplot , kde=True ) # kde=True in histplot adds a smooth curve to the histogram.
g.map_lower(sn.kdeplot) 
```
