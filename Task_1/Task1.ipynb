#### 3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]

```python
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

#### 4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP,Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.


```python
import pandas as pd
```

```python
pokemon_data = pd.read_csv('Pokemon.csv', encoding='latin')
pokemon_data.head(5)
```

```python
pokemon_data.size
```

```python
pokemon_data_heap = pokemon_data.iloc[: , 5:11]
pokemon_data_heap.head(5)
```

```python
pokemon_data_corr_pearson = pokemon_data_heap.corr(method = 'pearson')
pokemon_data_corr_pearson.head(5)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(pokemon_data_corr_pearson, annot=True, fmt=".4f", cmap = 'Blues')
plt.show()

# annot -- This will display the data values in each cell of the heatmap.
# fmt -- floating-point numbers
# cmap -- colormaps

```

```python
pokemon_data_corr_spearman = pokemon_data_heap.corr(method = 'spearman')
pokemon_data_corr_spearman.head(5)
```

```python
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(pokemon_data_corr_spearman, annot=True, fmt=".4f", cmap = 'Blues')
plt.show()
```
