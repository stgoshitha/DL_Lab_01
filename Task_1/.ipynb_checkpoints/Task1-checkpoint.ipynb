1. Create a random 4x4 array from an exponential distribution. Refer numpy functions.

```python
data = np.random.exponential(scale=1.0, size=(4,4))
print(data)
data.size
```

```python
array2 = np.random.exponential(scale=1.0, size=(100000,1))
print(array2)
```

```python
import matplotlib.pyplot as plt

plt.hist(array2, density=True, bins=100, color="blue",
label="rand", rwidth=0.8)
plt.axis([-1, 5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
import pandas as pd
```

```python
pokemon_data = pd.read_csv('Pokemon.csv', encoding='latin')
pokemon_data.head(5)
```

```python
pokemon_data.size
```

```python
pokemon_data_heap = pokemon_data.iloc[: , 5:11]
pokemon_data_heap.head(5)
```

```python
pokemon_data_corr = pokemon_data_heap.corr()
pokemon_data_corr.head(5)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(pokemon_data_corr, annot=True, fmt=".4f", cmap = 'Blues')
plt.show()

# annot -- This will display the data values in each cell of the heatmap.
# fmt -- floating-point numbers
# cmap -- colormaps

```
